```python
import json[{"ID": "TME", "Name": "Haven", "Organizer": "Admin Skye", "Start Date": "2022-01-25", "Start Time": "10:40:00", "End Date": "2022-01-25", "End Time": "14:40:00", "Users Registered": ["Brimstone Sharma"], "Capacity": 50, "Seats Available": 49}, {"ID": "YXE", "Name": "Ascent", "Organizer": "Admin Skye", "Start Date": "2021-12-19", "Start Time": "15:00:00", "End Date": "2021-12-19", "End Time": "15:50:00", "Users Registered": ["Brimstone Sharma", "Omen Kumar"], "Capacity": 10, "Seats Available": 8}, {"ID": "AW4", "Name": "IceBox", "Organizer": "Admin Cypher", "Start Date": "2022-02-10", "Start Time": "12:50:00", "End Date": "2022-02-10", "End Time": "13:40:00", "Users Registered": [], "Capacity": 100, "Seats Available": 100}]
import string
import random
from json import JSONDecodeError
from datetime import datetime,date

def AutoGenerate_EventID():
    #generate a random Event ID
    Event_ID=''.join(random.choices(string.ascii_uppercase+string.digits,k=3))
    return Event_ID

def Register(type,member_json_file,organizer_json_file,Full_Name,Email,Password):
    '''Register the member/ogranizer based on the type with the given details'''
    if type.lower()=='organizer':
        f=open(organizer_json_file,'r+')
        d={
            "Full Name":Full_Name,
            "Email":Email,
            "Password":Password
        }
        try:
            content=json.load(f)
            if d not in content:
                content.append(d)
                f.seek(0)
                f.truncate()
                json.dump(content,f)
        except JSONDecodeError:
            l=[]
            l.append(d)
            json.dump(l,f)
        f.close()
    else:
        f=open(member_json_file,'r+')
        d={
            "Full Name":Full_Name,
            "Email":Email,
            "Password":Password
        }
        try:
            content=json.load(f)
            if d not in content:
                content.append(d)
                f.seek(0)
                f.truncate()
                json.dump(content,f)
        except JSONDecodeError:
            l=[]
            l.append(d)
            json.dump(l,f)
        f.close()


def Login(type,members_json_file,organizers_json_file,Email,Password):
    '''Login Functionality || Return True if successful else False'''
    d=0
    if type.lower()=='organizer':
        f=open(organizers_json_file,'r+')
    else:
        f=open(members_json_file,'r+')
    try:
        content=json.load(f)
    except JSONDecodeError:
        f.close()
        return False
    for i in range(len(content)):
        if content[i]["Email"]==Email and content[i]["Password"]==Password:
            d=1
            break
    if d==0:
        f.close()
        return False
    f.close()
    return True

def Create_Event(org,events_json_file,Event_ID,Event_Name,Start_Date,Start_Time,End_Date,End_Time,Users_Registered,Capacity,Availability):
    '''Create an Event with the details entered by organizer'''
    event = {
        "ID": Event_ID,
        "Name": Event_Name,
        "Organizer": org,
        "Start Date": Start_Date,
        "Start Time": Start_Time,
        "End Date": End_Date,
        "End Time": End_Time,
        "Users Registered": Users_Registered,
        "Capacity": Capacity,
        "Seats Available": Availability
    }
    events_data = []

    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)
    except JSONDecodeError:
        pass

    events_data.append(event)

    with open(events_json_file, 'w') as f:
        json.dump(events_data, f, indent=4)

def View_Events(org,events_json_file):
    '''Return a list of all events created by the logged in organizer'''
    events_data = []
    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)
    except JSONDecodeError:
        pass

    organizer_events = [event for event in events_data if event["Organizer"] == org]
    return organizer_events

def View_Event_ByID(events_json_file,Event_ID):
    '''Return details of the event for the event ID entered by user'''
    events_data = []
    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)
    except JSONDecodeError:
        pass

    event = next((event for event in events_data if event["ID"] == Event_ID), None)
    return [event] if event else []

def Update_Event(org,events_json_file,event_id,detail_to_be_updated,updated_detail):
    '''Update Event by ID || Take the key name to be updated from member, then update the value entered by user for that key for the selected event
    || Return True if successful else False'''
    events_data = []
    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)
    except JSONDecodeError:
        pass

    event = next((event for event in events_data if event["ID"] == event_id), None)
    if not event:
        return False

    if detail_to_be_updated in event and detail_to_be_updated != "ID":
        event[detail_to_be_updated] = updated_detail
        with open(events_json_file, 'w') as f:
            json.dump(events_data, f, indent=4)
        return True

    return False
def Delete_Event(org,events_json_file,event_ID):
    '''Delete the Event with the entered Event ID || Return True if successful else False'''
    events_data = []
    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)
    except JSONDecodeError:
        pass

    event = next((event for event in events_data if event["ID"] == event_ID), None)
    if not event:
        return False

    events_data.remove(event)
    with open(events_json_file, 'w') as f:
        json.dump(events_data, f, indent=4)
    return True


def Register_for_Event(events_json_file,event_id,Full_Name):
    '''Register the logged in member in the event with the event ID entered by member. 
    (append Full Name inside the "Users Registered" list of the selected event)) 
    Return True if successful else return False'''
    date_today=str(date.today())
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    '''Write your code below this line'''
    events_data = []
    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)
    except JSONDecodeError:
        pass

    event = next((event for event in events_data if event["ID"] == event_id), None)
    if not event:
        return False

    if len(event["Users Registered"]) < event["Capacity"]:
        event["Users Registered"].append(Full_Name)
        event["Seats Available"] -= 1
        with open(events_json_file, 'w') as f:
            json.dump(events_data, f, indent=4)
        return True

    return False

def fetch_all_events(events_json_file,Full_Name,event_details,upcoming_ongoing):
    '''View Registered Events | Fetch a list of all events of the logged in memeber'''
    '''Append the details of all upcoming and ongoing events list based on the today's date/time and event's date/time'''
    date_today = str(date.today())
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    
    try:
        with open(events_json_file, 'r') as f:
            events_data = json.load(f)

        for event in events_data:
            event_start_datetime = datetime.strptime(f"{event['Start Date']} {event['Start Time']}", "%Y-%m-%d %H:%M:%S")
            event_end_datetime = datetime.strptime(f"{event['End Date']} {event['End Time']}", "%Y-%m-%d %H:%M:%S")

            if event_start_datetime <= now and event_end_datetime >= now:
                # Event is ongoing
                upcoming_ongoing.append(event)

            if event_start_datetime > now:
                # Event is upcoming
                event_details.append(event)

    except FileNotFoundError:
        print("Events file not found.")

    return event_details, upcoming_ongoing
def Update_Password(members_json_file, Full_Name, new_password):
    '''Update the password of the member by taking a new password || Return True if successful else return False'''
    members_data = []
    try:
        with open(members_json_file, 'r') as f:
            members_data = json.load(f)
    except JSONDecodeError:
        pass

    for member in members_data:
        if member["Full Name"] == Full_Name:
            member["Password"] = new_password
            with open(members_json_file, 'w') as f:
                json.dump(members_data, f, indent=4)
            return True

    return False

def View_all_events(events_json_file):
    '''Read all the events created | DO NOT change this function'''
    '''Already Implemented Helper Function'''
    details=[]
    f=open(events_json_file,'r')
    try:
        content=json.load(f)
        f.close()
    except JSONDecodeError:
        f.close()
        return details
    for i in range(len(content)):
        details.append(content[i])
    return details
```
